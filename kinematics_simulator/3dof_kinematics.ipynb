```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# end effecter
rx = 70
ry = 80
rz = 70
# link length
L1  = 70
L2  = 85
L3  = 85
```

```python
def forward_kinematics2(th1, th2):
    px = L2 * np.cos(th2) * np.cos(th1)
    py = L2 * np.cos(th2) * np.sin(th1)
    pz = L2 * np.sin(th2) + L1
    return px, py, pz
```

```python
def forward_kinematics3(th1, th2, th3):
    px = (L2 * np.cos(th2) + L3 * np.cos(th2 + th3)) * np.cos(th1)
    py = (L2 * np.cos(th2) + L3 * np.cos(th2 + th3)) * np.sin(th1)
    pz =  L2 * np.sin(th2) + L3 * np.sin(th2 + th3)  + L1
    return px, py, pz
```

```python
def inverse_kinematics1(x, y, z):
    OP    = np.sqrt(x**2 + y**2)
    alpha = np.arccos((L2**2 + L3**2 - OP**2) / (2 * L2 * L3))
    beta  = np.arccos((L2**2 + OP**2 - L3**2) / (2 * L2 * OP))
    gamma = np.arcsin((rz - L1) / OP)
    th1 = np.arctan2(y, x)
    th2 = gamma - beta
    th3 = np.pi - alpha
    return th1, th2, th3
```

```python
def inverse_kinematics2(x, y, z):
    OP    = np.sqrt(x**2 + y**2)
    alpha = np.arccos((L2**2 + L3**2 - OP**2) / (2 * L2 * L3))
    beta  = np.arccos((L2**2 + OP**2 - L3**2) / (2 * L2 * OP))
    gamma = np.arcsin((rz - L1) / OP)
    th1 = np.arctan2(y, x)
    th2 = gamma + beta
    th3 = -(np.pi - alpha)
    return th1, th2, th3
```

```python
th1, th2, th3 = inverse_kinematics1(rx, ry, rz)
print("th1=", th1, "(", np.degrees(th1), ")")
print("th2=", th2, "(", np.degrees(th2), ")")
print("th3=", th3, "(", np.degrees(th3), ")")
px1, py1, pz1 = [0, 0, L1]
px2, py2, pz2 = forward_kinematics2(th1, th2)
px3, py3, pz3 = forward_kinematics3(th1, th2, th3)
print("px=", px3)
print("py=", py3)
print("pz=", pz3)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
x = np.array([0, px1, px2, px3])
y = np.array([0, py1, py2, py3])
z = np.array([0, pz1, pz2, pz3])
ax.plot(x, y, z, marker="o")
plt.show()
```

```python
th1, th2, th3 = inverse_kinematics2(rx, ry, rz)
print("th1=", th1, "(", np.degrees(th1), ")")
print("th2=", th2, "(", np.degrees(th2), ")")
print("th3=", th3, "(", np.degrees(th3), ")")
px1, py1, pz1 = [0, 0, L1]
px2, py2, pz2 = forward_kinematics2(th1, th2)
px3, py3, pz3 = forward_kinematics3(th1, th2, th3)
print("px=", px3)
print("py=", py3)
print("pz=", pz3)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
x = np.array([0, px1, px2, px3])
y = np.array([0, py1, py2, py3])
z = np.array([0, pz1, pz2, pz3])
ax.plot(x, y, z, marker="o")
plt.show()
```
